با توجه به نمودار همبستگی بیشترین همبستگی مثبت با قیمت خانه را میانگین درآمد ماهیانه دارد و بیشترین همبستگی منفی هم مربوط به عرض جغرافیایی است و به ترتیب کمترین همبستگی با قیمت خانه مربوط متوسط تعداد اتاق خواب ها است که با افزایش قیمت خانه تعداد اتاق خواب ها به علت کاهش قدرت خرید مردم کاهش می یابد. همچمنین این افزایش قیمت سبب می شود که میانگین تعداد افراد کمتری در یک واحد مسکونی زندگی می کنند و در نهایت تعداد افراد کمتری در این مناطق زندگی کنن
